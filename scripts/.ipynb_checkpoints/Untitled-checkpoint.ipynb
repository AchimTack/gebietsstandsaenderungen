```python
import pandas as pd
from pathlib2 import Path
```

```python
def load_frame(filename):
    return pd.read_csv(str(filename),encoding='utf-8', sep='\t', decimal=',',converters={'ags_new':str,'ags_old':str})
```

```python
p = Path('../data/output/').glob('????.csv')
```

```python
n1, n2 = list(p)[0:2]
print n1, n2
```

```python
f1 = load_frame(n1)
f2 = load_frame(n2)
```

```python
pd.merge(f1, f2, how='outer', left_on='ags_old', right_on='ags_new')
```
