```python
import pandas as pd
from pathlib2 import Path
```

```python
def load_frame(filename):
    df = pd.read_csv(str(filename),encoding='utf-8', sep='\t', decimal=',',converters={'ags_new':str,'ags_old':str})[['ags_old','ags_new']]
    return df.set_index('ags_new')
```

```python
p = Path('../data/output/').glob('????.csv')
```

```python
n1, n2, n3 = list(p)[0:3]
print n1, n2, n3
```

```python
f1 = load_frame(n1)
f2 = load_frame(n2)
f3 = load_frame(n3)
```

```python
t = pd.merge(f1, f2, how='outer', right_on='ags_old', left_index=True, suffixes=['2010','2011'])
```

```python
t2 = pd.merge(t, f3, how='outer', right_on='ags_old', left_index=True, suffixes=['2011','2012'])
```
